```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/jerapah.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

print("Size of the image:")
print(img.shape)

print("Class of the image:")
print(img.dtype)

plt.figure(1)
plt.imshow(img)
plt.title("Figure 1: Image with Blue Lines")
y, x, _ = img.shape
plt.axhline(y=y//2, color='b', linewidth=2)
plt.axvline(x=x//2, color='b', linewidth=2)
plt.show()

img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
X, Y = np.meshgrid(range(img_gray.shape[1]), range(img_gray.shape[0]))

fig = plt.figure(2)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, img_gray, cmap='gray', edgecolor='none')
plt.colorbar(ax.plot_surface(X, Y, img_gray, cmap='gray', edgecolor='none'))
plt.title("Figure 2: 3D Plot of the Image")
plt.show()

x_crop, y_crop, width, height = 150, 100, 200, 200
cropped_img = img[y_crop:y_crop+height, x_crop:x_crop+width]

plt.figure(3)
plt.imshow(cropped_img)
print("Size Cropped Image:")
print(cropped_img.shape)
plt.title("Figure 3: Cropped Image")
plt.show()

plt.figure(4)
plt.imshow(img_gray, cmap='gray')
plt.title("Figure 4: Image Grayscale")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

eagle_image = cv2.imread('/content/eagle.jpg')
bird_image = cv2.imread('/content/bird.jpg')

eagle_image = cv2.cvtColor(eagle_image, cv2.COLOR_BGR2RGB)
bird_image = cv2.cvtColor(bird_image, cv2.COLOR_BGR2RGB)

if eagle_image.shape != bird_image.shape:
    bird_image = cv2.resize(bird_image, (eagle_image.shape[1], eagle_image.shape[0]))

print("Size of Eagle Image:", eagle_image.shape)
print("Size of Bird Image:", bird_image.shape)

plus_image = cv2.add(eagle_image, bird_image)

plt.figure(1)
plt.imshow(plus_image)
plt.title("Plus Image")
plt.axis('off')

rata = (eagle_image / 2 + bird_image / 2).astype(np.uint8)
rata_rata = ((eagle_image + bird_image) / 2).astype(np.uint8)

plt.figure(2)
plt.imshow(rata)
plt.title("Rata")
plt.axis('off')

plt.figure(3)
plt.imshow(rata_rata)
plt.title("Rata2")
plt.axis('off')

if eagle_image.shape[0] >= 200:
    plt.figure(4)
    plt.plot(plus_image[200, :, 0], label='Red Channel')
    plt.plot(plus_image[200, :, 1], label='Green Channel')
    plt.plot(plus_image[200, :, 2], label='Blue Channel')
    plt.title("Tambah (Baris ke-200)")
    plt.legend()
    plt.grid()

    plt.figure(5)
    plt.plot(rata_rata[200, :, 0], label='Red Channel')
    plt.plot(rata_rata[200, :, 1], label='Green Channel')
    plt.plot(rata_rata[200, :, 2], label='Blue Channel')
    plt.title("Rata2 (Baris ke-200)")
    plt.legend()
    plt.grid()

    plt.figure(6)
    plt.plot(rata[200, :, 0], label='Red Channel')
    plt.plot(rata[200, :, 1], label='Green Channel')
    plt.plot(rata[200, :, 2], label='Blue Channel')
    plt.title("Rata (Baris ke-200)")
    plt.legend()
    plt.grid()
else:
    print("Gambar terlalu kecil untuk memplot baris ke-200")

plt.show()
```
